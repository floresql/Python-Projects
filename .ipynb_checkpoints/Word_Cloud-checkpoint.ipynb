```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

# Replace the text below with the full text extracted from "Data & Reporting Analyst .pdf"
text = """
analyze data
SQL
data analysis 
reporting
data science
business intelligence 
financial analysis
SQL
data preparation
automation 
Alteryx
SSIS
Oracle
Excel
Python
Powershell
SAS
JavaScript
C++
data visualization
Tableau
Power BI
Microstrategy 
statistical analysis
"""

# Set up stopwords
stopwords = set(STOPWORDS)

# Generate the word cloud with a horizontal layout (prefer_horizontal=1)
wordcloud = WordCloud(width=800, height=200, background_color='white',
                      stopwords=stopwords, prefer_horizontal=0).generate(text)

# Display the word cloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
